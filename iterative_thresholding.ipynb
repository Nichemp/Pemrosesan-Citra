```python
import cv2
import numpy as np

def iterative_thresholding(image):
    # Langkah 1: Pilih nilai ambang awal (T0)
    T = 128  # atau np.mean(image)

    while True:
        # Langkah 2: Pisahkan area menjadi dua berdasarkan T
        area1 = image[image <= T]  # Area 1 (â‰¤ T)
        area2 = image[image > T]   # Area 2 (> T)

        # Langkah 3: Hitung nilai rata-rata dari area1 dan area2
        X1 = area1.mean() if len(area1) > 0 else 0  # Rata-rata intensitas di area1
        X2 = area2.mean() if len(area2) > 0 else 0  # Rata-rata intensitas di area2

        # Langkah 4: Hitung nilai ambang baru
        T_new = (X1 + X2) / 2

        # Langkah 5: Ulangi sampai T stabil
        if abs(T - T_new) < 1:  # Jika perbedaan nilai T sangat kecil, hentikan iterasi
            break

        T = T_new

    # Langkah 6: Kembalikan nilai ambang yang stabil
    return T

def main(image_path):
    # Membaca gambar grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Menerapkan algoritma iteratif thresholding
    threshold_value = iterative_thresholding(image)
    print(f'Nilai ambang yang ditentukan: {threshold_value}')

    # Menerapkan thresholding menggunakan nilai ambang yang ditentukan
    _, thresholded_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)

    # Menampilkan gambar asli dan hasil thresholding
    cv2.imshow('Hasil Thresholding', thresholded_image)

    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Ubah 'path_to_your_image.jpg' ke path gambar yang ingin diproses
if __name__ == "__main__":
    main('grayscale2.jpeg')
```
